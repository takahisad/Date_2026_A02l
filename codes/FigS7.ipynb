This notebook reproduces plots in Fig S7.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import scipy.stats as stats
import seaborn as sns
```

```python
#Fig S7C
crawl_before=np.load('../data/FigS7/FigS7C_data/240912_A02l_Flp_CsCh_crawl_before.npy')
crawl_stim=np.load('../data/FigS7/FigS7C_data/240912_A02l_Flp_CsCh_crawl_stim.npy')
fig=plt.figure(figsize=(3,5))
ax=fig.add_subplot(1,1,1)
ax = sns.swarmplot(data=[crawl_before, crawl_stim], color=".25", size=6)
plt.plot([np.zeros(len(crawl_before)), np.ones(len(crawl_stim))],
         [crawl_before, crawl_stim], c='k', lw=0.25)

x1, x2 = 0, 1
y, h, col = crawl_before.max() + 0.5, 0, 'k'
ax.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
ax.text((x1+x2)*.5, y+h+0.2, "n.s.", ha='center', va='bottom', color=col,  fontsize=20)
plt.ylim(-0.2,9)
plt.xlim(-0.5, 1.5)
plt.ylabel('# forward crawling in 5 sec', fontsize=20)
plt.gca().axes.xaxis.set_ticklabels([]) #delete numbers in xticks
plt.text(-0.18,-1.3, 'Light \n OFF', fontsize=14)
plt.text(-0.18+1,-1.3, 'Light \n ON', fontsize=14)
plt.gca().axes.xaxis.set_ticklabels([])
plt.yticks(fontsize=15)
plt.tight_layout()
plt.savefig('figures/FigS7.svg')
```

```python
#quantification of Fig S7C
s,p=stats.wilcoxon(crawl_before, crawl_stim)
print(f'p-value Fig S7C: {p}')
```

```python
#Fig S7E
crawl_before = np.load('../data/FigS7/FigS7E_data/250516_A02lsplit2_CsCh_crawl_before.npy')
crawl_stim = np.load('../data/FigS7/FigS7E_data/250516_A02lsplit2_CsCh_crawl_stim.npy')
fig=plt.figure(figsize=(3,5))
plt.rcParams["font.family"] = "Arial"
ax=fig.add_subplot(1,1,1)
ax = sns.swarmplot(data=[crawl_before, crawl_stim], color=".25", size=6)
plt.plot([np.zeros(len(crawl_before)), np.ones(len(crawl_stim))],
         [crawl_before, crawl_stim], c='k', lw=0.25)
plt.ylim(-0.1,9)
plt.xlim(-0.5, 1.5)
plt.gca().axes.xaxis.set_ticklabels([]) #delete numbers in xticks
plt.text(0,-1.3, 'Light\nOFF', fontsize=14, ha='center')
plt.text(1,-1.3, 'Light\nON', fontsize=14, ha='center')
plt.ylabel('# forward crawling in 5 sec', fontsize=20)
plt.yticks(fontsize=15)
x1, x2 = 0, 1
y, h, col = crawl_before.max() + 0.5, 0, 'k'
ax.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
ax.text((x1+x2)*.5, y+h-0.02, "***", ha='center', va='bottom', color=col,  fontsize=20)
plt.tight_layout()
plt.savefig('figures/FigS7E.svg')
```
